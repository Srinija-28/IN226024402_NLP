```python
#Task 1: Conceptual Understanding
#1. Difference between "Love" and "love"
#In Natural Language Processing, text is often case-sensitive. This means that words like "Love" and "love" are treated as different tokens unless normalization is applied. Converting text to lowercase ensures consistency and prevents duplication of meaning.

#2. What happens if stopwords are not removed?
#If stopwords (such as "is", "the", "and") are not removed, the dataset becomes noisy and less meaningful. These words occur frequently but contribute little to the actual meaning, which can reduce model efficiency and accuracy.

#3. When removing stopwords can be harmful
#Removing stopwords is not always beneficial. In some cases, it can distort meaning:
#In sentiment analysis, words like "not" are important (e.g., "not good" vs "good").
#In question answering systems, words like "what", "when", "where" are essential for understanding the query.

#4. Stemming vs Lemmatization
#Stemming reduces a word to its root form by cutting off prefixes or suffixes, often without considering meaning (e.g., "running" → "run").
#Lemmatization, on the other hand, converts words into their meaningful base form using vocabulary and grammar (e.g., "better" → "good"). Lemmatization is more accurate but computationally heavier.
```

```python
#Task 2:Build Advanced Preprocessing Function
import re
def preprocess_text(text):
    if not text or text.strip() == "":
        return [], ""
    # Remove URLs and emails
    text = re.sub(r'http\S+|www\S+|@\S+', '', text)
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    # Convert to lowercase
    text = text.lower()
    # Reduce repeated characters
    text = re.sub(r'(.)\1+', r'\1', text)
    # Remove special characters
    text = re.sub(r'[^a-z\s]', '', text)
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    # Tokenization
    tokens = text.split()
    # Remove short tokens (≤2), keep "no" and "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]
    # Form cleaned sentence
    clean_sentence = " ".join(tokens)
    return tokens, clean_sentence
```

```python
#Task 3: Stress Testing
test_sentences = [
"Get 100% FREE access now!!!",
"I absolutely looooved this product 😍😍",
"Worst service ever... 0/10",
"Call me at 9876543210",
"This is THE best course!!!",
"Visit https://openai.com now!",
"Nooooo this is baaad!!!",
"OK OK OK I got it",
"Win $$$ now!!! Limited offer!!!",
"I am not happy with this"
]
# Printing Output in Required Format
for i, sentence in enumerate(test_sentences, start=1):
    tokens, clean = preprocess_text(sentence)
    print(f"{i}. Original Text: {sentence}")
    print(f"   Cleaned Tokens: {tokens}")
    print(f"   Cleaned Sentence: {clean}")
    print("-" * 60)
```

```python
#Task 4: Token Analytics
# Function for Token Analytics
def token_analysis(tokens):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    if total_tokens == 0:
        avg_length = 0
    else:
        avg_length = sum(len(word) for word in tokens) / total_tokens

    return total_tokens, unique_tokens, avg_length
# Applying on all test sentences
for i, sentence in enumerate(test_sentences, start=1):
    tokens, clean = preprocess_text(sentence)
    total, unique, avg_len = token_analysis(tokens)
    print(f"{i}. Sentence: {clean}")
    print(f"   Total Tokens: {total}")
    print(f"   Unique Tokens: {unique}")
    print(f"   Average Token Length: {avg_len:.2f}")
    print("-" * 60)
```

```python
#Task 5: Frequency Analysis
from collections import Counter
# Collect all tokens from all sentences
all_tokens = []
for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)
# Count frequency
word_counts = Counter(all_tokens)
# Top 10 frequent words
top_10 = word_counts.most_common(10)
# Least 5 frequent words
least_5 = word_counts.most_common()[-5:]
print("Top 10 Most Frequent Words:")
for word, count in top_10:
    print(f"{word} : {count}")
print("\nTop 5 Least Frequent Words:")
for word, count in least_5:
    print(f"{word} : {count}")
```

```python
#Task 6: Build Full Pipeline
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
result = full_pipeline(test_sentences)
print("All Tokens:")
print(result["tokens"])
print("\nClean Sentences:")
print(result["clean_sentences"])    
```
